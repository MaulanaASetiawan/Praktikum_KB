### Import Library

```python
import pandas as pd
import prettytable as ptl
```

### Import Dataset yang sudah di acc

```python
df = pd.read_csv("Steel_industry_data.csv")
```

### Menampilkan informasi dari dataset

```python
df.info()
```

### Menampilkan statistik dari dataset yang bertipe Numerik

```python
df.describe()
```

### Menampilkan 10 record pertama dari dataset dengan attribute bertipe angka.

```python
df.select_dtypes(include=['int64','float64']).head(10)
```

### Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka, dan dimasukkan kedalam Dictionary.

```python
datasetDictionary = {
    'Jumlah Record': df.shape[0],
    'Jumlah Attribute': df.shape[1],
    'Nama Attribute': df.columns.tolist(),
    'Jumlah Attribute Dengan Tipe Angka': df.select_dtypes(include=['int64', 'float64']).shape[1],
}

table = ptl.PrettyTable()
table.add_column('Jumlah Record', [datasetDictionary['Jumlah Record']])
table.add_column('Jumlah Attribute', [datasetDictionary['Jumlah Attribute']])
table.add_column('Nama Attribute', ['\n'.join(datasetDictionary['Nama Attribute'])])
table.add_column('Jumlah Attribute Dengan Tipe Angka', [datasetDictionary['Jumlah Attribute Dengan Tipe Angka']])
print(table)

```

### Menampilkan nilai dataset pada baris dimulai dengan 3 digit NIM akhir hingga data terakhir dan kolom ke 4 hingga kolom terakhir menggunakan ke 2 metode slice yaitu (iloc dan loc)

```python
df.iloc[24:, 3:]
```

```python
df.loc[24:, datasetDictionary['Nama Attribute'][3:]]
```
